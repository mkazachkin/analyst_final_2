# Задача 3
Мы хотим провести A/B-тест для трех источников траффика. Нынешняя конверсия равна 5%, мы ожидаем прирост в 0.2%. Уровень доверия 97% и уровень мощности 87%. Всего на наш продукт заходит 40000 пользователей в месяц.

## Задача 3.1
За сколько дней мы сможем протестировать гипотезу? Что вы можете посоветовать по результатам подсчета?

# Решение
По условию задачи у нас:
* Количество вариантов (источников траффика, шт.): 3
* Средний показатель (конверсия, %): 5

* Достоверность (уровень доверия, %): 97
* Мощность (уровен мощности, %): 87

Для рассчета размера выборки воспользуемся онлайн калькулятором достоверности A/B-тестов на сайте [mindbox.ru](https://mindbox.ru/tools/ab-test-calculator/).

Общий объем выборки должен составлять 1522706 (один миллион пятьсот двадцать две тысячи семьсот шесть) пользователей. Рассчитаем количество дней, которое потребуется для проведения A/B-тестирования при средней посещаемости сайта в 40000 человек в месяц.

```python
int(1_522_706 / 40_000) + 1
```

## Ответ:
Для проведение A/B-тестирования потребуется не менее 39 дней. Целесообразность проведение столь длительного теста ради увеличения конверсии на 4% в относительном выражении при такой посещаемости сайта определяется финансовыми показателями проекта. Если потенциальная финансовая выгода от увеличения конверсии значительно превысит стоимость тестирования, то это имеет смысл. В противном случае, возможно имеет смысл поискать другие источники траффика, которые бы дали больший прирост конверсии.

Если исходить из заданных параметров мощности и достоверности, то нужно рассматривать варианты источников траффика, которые могут увеличить конверсию минимум на 10% в относиельном выражении (0.5% в абсолютном), в этом случае на проведение A/B-тестирования потребуется значительно меньше времени (до 7 дней).

Если же ожидаемый прирост из других источников траффика не превышает 0.2%, то для снижения потребного времени придется жертвовать мощностью и достоверностью A/B-тестирования. Но с учетом низкого ожидаемого эффекта, значительно повышается вероятность ошибки. При достоверности в 95% и мощности 60% на проведение потребуется как минимум 22 дня.
